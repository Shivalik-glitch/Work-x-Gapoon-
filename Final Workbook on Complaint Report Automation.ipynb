```python
def same_next1(X) :
    first_day = X[0].day - 1
    list3 = []
    for i in X :
        if (i.day == first_day and i.hour >= 19 and i.hour <= 23 and i.minute <= 59) :
            i = 'Next Day'
        elif (i.day != first_day and i.hour >= 0 and i.hour == 8 and i.minute <= 59) :
            i = 'Same Day Before 10'
        else :
            i = 'Same day before 1 hour'
        list3.append(i)
    return list3
```

```python
def Complaint(Y) :
    first_day = Y[0].day - 1
    list4 = []
    for i in Y :
        if (i.day == first_day and i.hour >= 19 and i.hour <= 23 and i.minute <= 59) :
            i = 'Complaint'
        elif (i.day != first_day and i.hour >= 0 and i.hour <= 8 and i.minute <= 59) :
            i = 'Complaint'
        elif (i.day != first_day and i.hour >= 9 and i.hour <= 10 ) :
            i = 'Complaint and within 1 hour'
        elif (i.day != first_day and i.hour >=10 and i.hour <= 11) :
            i = 'Non Complaint but more than 1 hour but less than 2' 
        else :
            i = 'Non Complaint and more than 1 hour'
        list4.append(i)
    return list4 
```
